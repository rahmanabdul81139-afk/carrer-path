```python
system_prompt = """
You are **CareerMentorAI**, an advanced career guidance agent built for students in schools and colleges.
Your role:
- Understand user goals, strengths, interests.
- Recommend career paths, skills, certifications, internships, and roadmaps.
- Provide motivational guidance and explain complex topics simply.
- Give structured, accurate, real-world career advice.
- Be friendly, supportive, and clear.

You MUST respond:
- In short paragraphs
- With bullet points when needed
- With action steps

NEVER provide harmful, medical, or illegal advice.
"""
```

```python
def ask_career_agent(user_message):
    # Combine system + user messages together (v1beta requirement)
    full_prompt = f"{system_prompt}\n\nUser: {user_message}\nAgent:"

    response = model.generate_content(full_prompt)
    return response.text

```

```python
ask_career_agent("I am a student who wants to become a cloud engineer. What should I do?")
```

```python
def chat_with_agent():
    print("AI Career Mentor Agent is ready!")
    print("Type 'exit' to stop.\n")

    while True:
        user_input = input("You: ")

        if user_input.lower() in ["exit", "quit", "stop"]:
            print("Agent: Goodbye! Keep learning and shining! ‚ú®")
            break

        reply = ask_career_agent(user_input)
        print("\nAgent:", reply, "\n")
```

```python
chat_with_agent()
```

```python
def generate_career_roadmap(career):
    prompt = f"""
{system_prompt}

User wants a step-by-step roadmap to become: {career}

Generate a clean, structured roadmap with:
- Required skills
- Certifications
- Projects
- Internships
- Timeline (0‚Äì12 months)
"""
    response = model.generate_content(prompt)
    return response.text
```

```python
def personality_test(answers):
    prompt = f"""
{system_prompt}

The user has answered the following personality test questions:

{answers}

Based on this, provide:
- Strengths
- Career type match
- Suggested roles
- Skill-building plan
"""
    response = model.generate_content(prompt)
    return response.text
```

```python
personality_scores = {
    "Logical Thinking": 5,
    "People Orientation": 3,
    "Structure Preference": 5,
    "Technology Interest": 5,
    "Repetitiveness Tolerance": 1,
    "Learning Motivation": 5
}

personality_scores
```

```python
import matplotlib.pyplot as plt

labels = list(personality_scores.keys())
values = list(personality_scores.values())

plt.figure(figsize=(10,5))
plt.bar(labels, values)
plt.xticks(rotation=45)
plt.title("Personality Trait Score Visualization")
plt.ylabel("Score (1‚Äì5)")
plt.grid(axis='y')
plt.show()
```
